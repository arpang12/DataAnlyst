```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0 , 10.0)
%matplotlib inline
```

```python
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head(30)
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
n = len(X)
```

```python
nam = 0
deno =0
for i in range(n):
    nam += (X[i] - mean_x) * (Y[i] - mean_y)
    deno += (X[i] -mean_x) ** 2
b1 = nam/deno
b0 = mean_y - (b1 * mean_x)
print(b1, b0)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)

y = b0 + b1 * x

plt.plot(x, y, color ='#13060d', label='Regression line')

plt.scatter(X, Y, c='#ffc34d', label='Scatter plot')

plt.xlabel('Head size')

plt.ylabel('Brain Wieght')

plt.legend()

plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range (n):
    y_pred = b0 + b1 *X[i]
    ss_t +=(Y[i] - mean_y) * 2
    ss_r +=(Y[i] - y_pred) * 2
ro= 1 -(ss_r/ss_t)
print(ro)
```
